# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Dataset

```python
dataset = pd.read_csv('BigBasketProject.csv', header = None)
transactions = []

for i in range(0,7219):
  transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
print(dataset)
```

# Apriori Training on Dataset


```python
from apyori import apriori
basket_intelligence = apriori(transactions = transactions, min_support = 0.002, min_confidence = 0.2, min_lift= 3, min_length = 2 )
```

# Visualizing

# Raw Results

```python
results = list (basket_intelligence)
results
```

# Display Format

```python
def inspect (results):
  product1    = [tuple(result[2][0][0])[0] for result in results]
  product2    = [tuple(result[2][0][1])[0] for result in results]
  supports =     [result[1] for result in results]
  confidences = [result[2][0][2] for result in results]
  lifts =       [result[2][0][3] for result in results]
  return list(zip(product1, product2, supports, confidences, lifts))

```

```python
from google.colab import data_table
data_table.enable_dataframe_formatter()
dataframe_intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product2', 'Support', 'Confidence', 'Lift'])
dataframe_intelligence
```
